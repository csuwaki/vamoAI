```python
import numpy as np
import statistics
from scipy import stats
```

```python
vendas = np.random.randint(size = (12,30), low = 0, high = 200)
print(vendas)
```

```python
media = np.mean(vendas)
media_arr = np.round_(media)
print(media_arr)
```

```python
mediana = np.median(vendas)
print(mediana)
```

```python
moda = stats.mode(vendas, axis = None)
print(moda)
```

```python
media_mensal = vendas.mean(axis=1)
print(media_mensal)
```

```python
mediana_men = np.median(vendas, axis = 1)
print(mediana_men)
```

```python
import matplotlib.pyplot as pl
```

```python
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro','Novembro', 'Dezembro']
```

```python
plt.figure(figsize=[12, 10])

plt.plot(meses, media_mensal, 'k:', color='navy') 
plt.title("Media de vendas mensal")

plt.grid(True)
plt.xlabel("Mês")
plt.ylabel("Média de vendas mensal")
plt.show()
```
